```r
rowPaste <- function(X) apply( X,1,paste, collapse=" "  )
tr[, label := rowPaste( tr[,c("LabelName1","RelationshipLabel","LabelName2"),with=F]  )  ]
dt <- tr[, .N , keyby=c("ImageID")]

dt <- tr[, list(
  XMin1=median(XMin1),
  XMax1=median(XMax1),
  YMin1=median(YMin1),
  YMax1=median(YMax1),
  XMin2=median(XMin2),
  XMax2=median(XMax2),
  YMin2=median(YMin2),
  YMax2=median(YMax2),
  N=.N
) , keyby=c("LabelName1","RelationshipLabel","LabelName2")  ]

setorderv( dt, "N",-1  )

head(dt,300)
```
